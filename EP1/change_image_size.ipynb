```python
import numpy as np
import pandas as pd
from PIL import Image
import PIL.ExifTags
from glob import glob
```

```python
def get_orientation(image):
    return {
        PIL.ExifTags.TAGS[k]: v
        for k, v in image._getexif().items()
        if k in PIL.ExifTags.TAGS
    }['Orientation']

def get_correct_oriented_image(path):
    img = Image.open(path)
    orientation = get_orientation(img)
    arr = np.array(img)
    
    number_rotations = {6: 3, 8:1}
    if orientation != 1:
        arr = np.rot90(arr, number_rotations[orientation])
        
    return arr
```

```python
for path in glob('/Users/gabimelo/Downloads/EP1_dataset/*.jpg'):
    image = Image.fromarray(get_correct_oriented_image(path))
    image = image.resize((3000, 4000))
    image.save('/Users/gabimelo/Downloads/resized_dataset/' + path.split('/')[-1])
```
