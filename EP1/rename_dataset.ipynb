```python
import shutil as sh
import pandas as pd

# cria diretório para o dataset das fotos renomeadas

!mkdir dataset
```

```python
# cria função para copiar um arquivo de foto e renomea-lo de acordo com as colunas da tabela de metadados
# (poderia só sobrescrecer, mas por cautela achei melhor copiar e guardar numa pasta diferente)

def copy_with_new_name(row):
    new_name = f"{row['classe']}_{row['exemplo']}_{row['periodo']}_{row['local']}_{row['fundo']}_{row['repeticao']}.jpg"
    sh.copyfile('original_dataset/' + row['arquivo'] + '.jpg', 'dataset/' + new_name)
    return new_name
```

```python
# lê o arquivo excel com a tabela de metadados

metadata = pd.read_excel('original_dataset/original_metadata.xlsx')
```

```python
# aplica a função sobre os arquivos originais de acordo com a coluna 'arquivo'

metadata['arquivo'] = metadata.apply(copy_with_new_name, axis=1)
```

```python
# cria uma nova tabela excel com os novos nomes de arquivo

metadata.to_excel('./metadata.xlsx', index=False)
```
